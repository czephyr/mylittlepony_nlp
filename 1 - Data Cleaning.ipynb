```python
len(df_complete)
```

```python
chr_count = df_complete["pony"].value_counts().head(30)
chr_count
```

```python
multiplier_1 = chr_count.sum()/len(df_complete)
```

```python
chr_count.sort_values(ascending=False)
```

```python
plt.grid(True, "both", linestyle="--")
plt.plot(range(1,1+len(chr_count)), multiplier_1*round(chr_count.cumsum()/chr_count.sum(),2))
plt.axhline(y=0.7, color='r', linestyle='--')
plt.axvline(x=12, color='r', linestyle='--')
plt.title("Cumulative Percentage of Dataset vs Sorted Number of Characters")
plt.xlabel("Number of Characters (sorted in descending order of dialogue)")
plt.ylabel("Percentage of Dataset")
plt.xticks(ticks = range(1,1+len(chr_count)), labels=chr_count.index, rotation=90)
plt.show()
```
